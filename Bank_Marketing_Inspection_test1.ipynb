# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```


## Check for missing values and data types



```python
# check Data types
print(df.dtypes)
```

```python
# Check for Missing Values
print(df.isnull().sum())
```

```python
# Get a Quick Summary
print(df.info())
```


## Summary statistics



```python
print(df.describe())
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
```

```python
print(df.columns)
```

```python
sns.countplot(data=df, x='y')
plt.title('Distribution of Target Variable (y)')
plt.xlabel('Target (y)')
plt.ylabel('Count')
plt.show()
```


# Correlation matrix for numerical features



```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numerical_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
#  Insights Gained
# 1. Customer behavior patterns can be studied (e.g., job, age, housing).

# 2. Economic indicators like euribor3m have strong influence on client decisions.

# 3. Imbalanced data reveals that predicting a "yes" might be challenging and needs special treatment in machine learning models.

# 4. This EDA (Exploratory Data Analysis) builds a strong foundation for predictive modeling (e.g., logistic regression, decision trees).
```


## Conclusion
